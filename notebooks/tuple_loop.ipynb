# Tuple - Immutable
Tuple Data type is exactly same as list data type except that it is immutable. i.e we cannot change values.


Tuple elements can be represented within parenthesis

```python
lst = [] # square bracket
tpl = () # parenthesis
```

```python
tpl1 = (1,3,4.2,"hello")
tpl2 = tpl1
print(id(tpl1)==id(tpl2)) # both tpl1, tpl2 point to same id of tuple
```

```python
tpl1[0] = 45 # TypeError: 'tuple' object does not support item assignment
```

```python
tpl1.count(2)
```

```python
tpl1.index(1)
```

```python
tpl = [2,4,5,1,3.4] # only int and float req
sorted(tpl)
```

```python
while_lp = (21,32,45,65,67,33,431,56,75,78,43,21,65,75)
i = 0
while i < len(while_lp):
    print("index=> {0} - value=> {1}".format(i, while_lp[i]))
    
    if i == 10:
        break
    
    i+=1
    if i == 5:
        continue
print("final count => ",i)
```
